Dashboard interativo dos resultados do pipeline ETL.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path

# Carrega resultados
df = pd.read_csv('../data/output/clientes_enriquecidos.csv')
df.head()
```

```python
# Distribuição por produto
fig = px.histogram(df, x='produto', title='Distribuição de Produtos')
fig.show()
```

```python
# Comprimento das mensagens
df['tamanho_mensagem'] = df['mensagem'].str.len()
fig = px.box(df, x='produto', y='tamanho_mensagem', title='Tamanho Mensagens por Produto')
fig.show()
```

```python
# Amostra mensagens
print('Exemplos de mensagens geradas:')
for i, msg in enumerate(df['mensagem'].head()):
    print(f'{i+1}. {msg[:100]}...')
```
