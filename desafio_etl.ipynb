# Desafio DIO - ETL com Python

Este notebook resolve o desafio usando a alternativa sem dependência de API.

Fluxo:
1. Extração: lista de clientes criada em Python.
2. Transformação: geração de mensagens personalizadas.
3. Carregamento: salvamento dos dados enriquecidos em CSV.

```python
from pathlib import Path
import csv

clientes = [
    {
        "nome": "Ana Souza",
        "conta": "0001-25",
        "cartao": "**** **** **** 1122",
        "produto": "cartão de crédito premium",
    },
    {
        "nome": "Bruno Lima",
        "conta": "0002-31",
        "cartao": "**** **** **** 3344",
        "produto": "empréstimo pessoal",
    },
    {
        "nome": "Carla Mendes",
        "conta": "0003-48",
        "cartao": "**** **** **** 5566",
        "produto": "investimentos automatizados",
    },
]

clientes
```

## Transformação

Nesta etapa criamos uma mensagem personalizada para cada cliente.

```python
def gerar_mensagem(cliente: dict) -> str:
    return (
        f"Olá, {cliente['nome']}! "
        f"Identificamos uma oportunidade para o seu perfil: {cliente['produto']}. "
        f"Conta {cliente['conta']} | Cartão {cliente['cartao']}."
    )

clientes_enriquecidos = []

for cliente in clientes:
    item = cliente.copy()
    item['mensagem'] = gerar_mensagem(cliente)
    clientes_enriquecidos.append(item)

clientes_enriquecidos
```

## Carregamento

Os dados processados são salvos em um arquivo CSV para facilitar a entrega do projeto.

```python
saida = Path("clientes_enriquecidos.csv")
campos = ["nome", "conta", "cartao", "produto", "mensagem"]

with saida.open("w", newline="", encoding="utf-8") as arquivo_csv:
    writer = csv.DictWriter(arquivo_csv, fieldnames=campos)
    writer.writeheader()
    writer.writerows(clientes_enriquecidos)

print(f"Arquivo gerado: {saida.resolve()}")
```

## Resultado

O notebook mostra o fluxo ETL completo sem depender da API externa, mantendo o foco no uso de Python para tratar e carregar os dados.
